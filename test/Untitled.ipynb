```python
import numpy as np
import matplotlib.pyplot as plt
import shelve
from scipy import sparse
from scipy.sparse import linalg as spla
%matplotlib notebook
```

```python
matrices = shelve.open('matrices.dat', 'r')

def plot_matrix(M):
    plt.imshow(np.log10(np.abs(M.A)))
```

```python
M = matrices['LHS'][5]

plt.figure()
plot_matrix(M)
plt.colorbar()
```

```python
LU = spla.splu(M, permc_spec='NATURAL')
```

```python
plt.figure()
plot_matrix(LU.U)
```
